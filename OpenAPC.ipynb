```python
## erst einmal ein package.modul (urllib,request) herunterladen und Pandas, das mit pd abgek√ºrzt wird
import urllib.request
import pandas as pd
```

```python
## URL wird in einen String "" gepackt, genauso der neue Dateiname
open_apc_url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
apc_file = "openapc.csv"
```

```python
## package.modul.funktion (die sich im Modul befindet)
urllib.request.urlretrieve(open_apc_url, apc_file)
```

```python
## Variable vergeben
apcs = pd.read_csv(apc_file)
```

```python
## anzeigen der Variablen = Inhalt der csv-Datei, die ersten und die letzten Zeilen werden angezeigt
apcs
```
