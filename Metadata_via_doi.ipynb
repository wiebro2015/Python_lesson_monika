# Übung
Title, short-container-title und Doi von 4 verschiedenen Publikationen ausgeben

```python
## importieren von 4 DOIs mit einer baseURL
## zunächst die Variable dois mit den 4 doi anlegen
## dann base_url definieren
## zur Vereinfachung die Variable aus base_url + doi anlegen
## in doi_json_data die Datein einlesen
## um die Auswahl anzeigen zu lassen, die Variable doi_data durch json.loads befüllen lassen indem mit print die
## einzelnen keys eingelesen werden
## mit print("\n") wird zwischen den einzelnen Angaben eine leere Zeile gesetzt

import urllib.request
import json

dois = ["10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.1038/35057062",
        "10.21105/joss.01006"]
base_url = "https://api.crossref.org/works/"

for doi in dois:
    full_url = base_url + doi
    doi_json_data = urllib.request.urlopen(full_url).read()
    print(doi_json_data)
    doi_data = json.loads(doi_json_data)
    print(doi_data["message"]["title"][0])
    print(doi_data["message"]["short-container-title"][0])
    print(doi)
    print("\n")
```
